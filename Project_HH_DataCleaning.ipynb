Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience_inmonth(str):
    
    no_ex = 'Не указано'
    
    if no_ex in str: 
        return np.nan
    
    months = ['месяц','месяца','месяцев']
    years = ['год','года','лет']
    
    str = str.replace('Опыт работы','')
    str_list = str[:20].split()
    
    experience = 0
    
    for month in months:
        if month in str_list:
            experience += int(str_list[str_list.index(month)-1])
            break
    
    for year in years:
        if year in str_list:
            experience += int(str_list[0])*12
            break
    
    return experience

hh_df['Опыт работы'].fillna('Не указано',inplace=True)
hh_df[hh_df['Опыт работы']=='Не указано']['Опыт работы'].value_counts()

hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience_inmonth)

print(hh_df[['Опыт работы','Опыт работы (месяц)']].tail(20))

hh_df.drop(labels='Опыт работы',axis=1,inplace=True)
hh_df['Опыт работы (месяц)'].isnull().value_counts()
```

```python
def get_city(str):
   
   million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
   capitals = ['Москва', 'Санкт-Петербург']
   
   city = str.split(',')[0]
   city = city.replace(' ','')
   
   if city in million_cities:
      return 'город-миллионник'
   elif city in capitals:
      return city
   else:
      return 'другие'
   
def move_ready(str):
    
    move_list = ['не готов к переезду', 'не готова к переезду', 'готов к переезду', 'готова к переезду', 'хочу переехать']
    
    for i, item in enumerate(move_list):
        if (item in str) and (i<2):
            return False
        if (item in str) and (i>=2):
            return True
    return False
        
def trip_ready(str):
    
    pos = str.find('командировк')
    
    if pos>-1:
        str_trip = str[str.rfind(',',0,pos):pos+11]
        
        if 'не готов' in str_trip:
            return False
        else:
            return True
    else:
        return False
    
   
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(move_ready)
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(trip_ready)
hh_df.drop('Город, переезд, командировки',axis=1,inplace=True)

print(hh_df['Город'].value_counts(normalize=True)*100)
print(hh_df[['Готовность к переезду','Готовность к командировкам']].value_counts(normalize=True))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def employment_encoding(data, feature, indicators):
    
    res_list = list()
    for i in range(len(indicators)):
        res_list.append(list())
    
    get_indicator = lambda x, str: True if x in str else False
    end = data[feature].shape[0]

    for row in range(end):
        str = data[feature].iloc[row]
        for i, item in enumerate(indicators):
            res_list[i].append(get_indicator(item,str))
        
    res_s = list()
    for i in range(len(res_list)):
        res_s.append(pd.Series(data=res_list[i]))
    
    return tuple(res_s)

indicators = ['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']
hh_df['полная занятость'], hh_df['частичная занятость'], hh_df['проектная работа'], hh_df['стажировка'], hh_df['волонтерство'] = employment_encoding(hh_df, 'Занятость', indicators)

indicators = ['полный день', 'гибкий график', 'сменный график', 'удаленная работа', 'вахтовый метод']
hh_df['полный день'], hh_df['гибкий график'], hh_df['сменный график'], hh_df['удаленная работа'], hh_df['вахтовый метод'] = employment_encoding(hh_df, 'График', indicators)

display(hh_df[['проектная работа','волонтерство']].value_counts())
display(hh_df[hh_df['вахтовый метод'] & hh_df['гибкий график']].shape[0])

hh_df.drop(labels=['Занятость','График'] ,axis=1, inplace=True)
```

```python
# Приводим дату к формату Datetime
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'],dayfirst=True)
hh_df['date'] = hh_df['Обновление резюме']

# Приводим название валюты к коду
def get_currency_id(item):
    
    curr_in_df = ['руб.','KZT','USD','бел.руб.','EUR','грн.','сум','KGZ','AZN']
    currency = ['RUB','KZT','USD','BYN','EUR','UAH','UZS','KGZ','AZN']
    try:
        i = curr_in_df.index(item)
        return currency[i]
    except ValueError:
        return None

hh_df['currency'] = hh_df['ЗП'].apply(lambda x: x.split()[1].replace(' ',''))
hh_df['currency'] = hh_df['currency'].apply(get_currency_id)

# Создаем столбец с величиной зарплаты формата float
hh_df['ЗП (руб)'] = hh_df['ЗП'].apply(lambda x: float(x.split()[0].replace(' ','')))

# Считываем файл с курсами фалют и приводим дату к формату datetime
curr_df = pd.read_csv('data/ExchangeRates.csv')
curr_df['date'] = pd.to_datetime(curr_df['date'])

# Проводим слияние баз
hh_df = hh_df.merge(curr_df, on=['currency','date'], how='left')

# Переводим все зарплаты в рубли и удаляем лишние столбцы
values = {
    'close': 1,
    'proportion': 1
}
hh_df = hh_df.fillna(values)

hh_df['ЗП (руб)'] = hh_df['ЗП (руб)']*hh_df['close']/hh_df['proportion']

# Удаляем все лишние столбцы
hh_df.drop(['per','time','close','vol','proportion','date','currency','ЗП'], axis=1, inplace=True)

display(hh_df.shape)
display(hh_df['ЗП (руб)'].median())
```
